```python
import pandas as pd
import numpy as np
```

```python
# clean for hdi
data = pd.read_csv("final dataset/HDR12_19.csv")

data_clean = data[['Country','le_2012','le_2013','le_2014','le_2015',
                   'le_2016','le_2017','le_2018','le_2019','le_2020',
                   'eys_2012','eys_2013','eys_2014','eys_2015','eys_2016',
                   'eys_2017','eys_2018','eys_2019','eys_2020','mys_2012',
                   'mys_2013','mys_2014','mys_2015','mys_2016','mys_2017',
                   'mys_2018','mys_2019','mys_2020','gnipc_2012','gnipc_2013',
                   'gnipc_2014','gnipc_2015','gnipc_2016','gnipc_2017','gnipc_2018',
                   'gnipc_2019','gnipc_2020']]

data_clean = data_clean.dropna()

dt_2012 = data_clean[['Country','le_2012','eys_2012','mys_2012','gnipc_2012']]
dt_2012['year'] = 2012
dt_2012.rename(columns = {'le_2012':'Life expectancy at birth', 'eys_2012':'Expected years of schooling',
                          'mys_2012':'Mean years of schooling', 'gnipc_2012':'GNI per capita'}, inplace = True)

dt_2013 = data_clean[['Country','le_2013','eys_2013','mys_2013','gnipc_2013']]
dt_2013['year'] = 2013
dt_2013.rename(columns = {'le_2013':'Life expectancy at birth', 'eys_2013':'Expected years of schooling',
                          'mys_2013':'Mean years of schooling', 'gnipc_2013':'GNI per capita'}, inplace = True)

dt_2014 = data_clean[['Country','le_2014','eys_2014','mys_2014','gnipc_2014']]
dt_2014['year'] = 2014
dt_2014.rename(columns = {'le_2014':'Life expectancy at birth', 'eys_2014':'Expected years of schooling',
                          'mys_2014':'Mean years of schooling', 'gnipc_2014':'GNI per capita'}, inplace = True)

dt_2015 = data_clean[['Country','le_2015','eys_2015','mys_2015','gnipc_2015']]
dt_2015['year'] = 2015
dt_2015.rename(columns = {'le_2015':'Life expectancy at birth', 'eys_2015':'Expected years of schooling',
                          'mys_2015':'Mean years of schooling', 'gnipc_2015':'GNI per capita'}, inplace = True)

dt_2016 = data_clean[['Country','le_2016','eys_2016','mys_2016','gnipc_2016']]
dt_2016['year'] = 2016
dt_2016.rename(columns = {'le_2016':'Life expectancy at birth', 'eys_2016':'Expected years of schooling',
                          'mys_2016':'Mean years of schooling', 'gnipc_2016':'GNI per capita'}, inplace = True)

dt_2017 = data_clean[['Country','le_2017','eys_2017','mys_2017','gnipc_2017']]
dt_2017['year'] = 2017
dt_2017.rename(columns = {'le_2017':'Life expectancy at birth', 'eys_2017':'Expected years of schooling',
                          'mys_2017':'Mean years of schooling', 'gnipc_2017':'GNI per capita'}, inplace = True)

dt_2018 = data_clean[['Country','le_2018','eys_2018','mys_2018','gnipc_2018']]
dt_2018['year'] = 2018
dt_2018.rename(columns = {'le_2018':'Life expectancy at birth', 'eys_2018':'Expected years of schooling',
                          'mys_2018':'Mean years of schooling', 'gnipc_2018':'GNI per capita'}, inplace = True)

dt_2019 = data_clean[['Country','le_2019','eys_2019','mys_2019','gnipc_2019']]
dt_2019['year'] = 2019
dt_2019.rename(columns = {'le_2019':'Life expectancy at birth', 'eys_2019':'Expected years of schooling',
                          'mys_2019':'Mean years of schooling', 'gnipc_2019':'GNI per capita'}, inplace = True)

dt_2020 = data_clean[['Country','le_2020','eys_2020','mys_2020','gnipc_2020']]
dt_2020['year'] = 2020
dt_2020.rename(columns = {'le_2020':'Life expectancy at birth', 'eys_2020':'Expected years of schooling',
                          'mys_2020':'Mean years of schooling', 'gnipc_2020':'GNI per capita'}, inplace = True)

dt_all = pd.concat([dt_2012, dt_2013, dt_2014, dt_2015, dt_2016, dt_2017, dt_2018, dt_2019, dt_2020])
```

```python
#read the datasets of exports + net_loans/borrowing
exports = pd.read_csv("final dataset/Volume of exports of goods and services.csv", header = 0)
net_loans = pd.read_csv("final dataset/General government primary net lending:borrowing.csv", header = 0)
# reconstruct four dataset into one dataset
data_2012 = []
country = exports['Country'].tolist()
ex_2012 = exports['2012'].tolist()
net_2012 = net_loans['2012'].tolist()
for i in range(len(country)):
    data_2012.append([country[i], ex_2012[i], net_2012[i], 2012])
data_2012 = pd.DataFrame(data_2012, columns=['Country', 'exports', 'net_loans', 'year'])

data_2013 = []
ex_2013 = exports['2013'].tolist()
net_2013 = net_loans['2013'].tolist()
for i in range(len(country)):
    data_2013.append([country[i],ex_2013[i], net_2013[i], 2013])
data_2013 = pd.DataFrame(data_2013, columns=['Country', 'exports', 'net_loans', 'year'])

data_2014 = []
ex_2014 = exports['2014'].tolist()
net_2014 = net_loans['2014'].tolist()
for i in range(len(country)):
    data_2014.append([country[i],ex_2014[i],  net_2014[i], 2014])
data_2014 = pd.DataFrame(data_2014, columns=['Country',  'exports', 'net_loans', 'year'])

data_2015 = []

ex_2015 = exports['2015'].tolist()
net_2015 = net_loans['2015'].tolist()
for i in range(len(country)):
    data_2015.append([country[i],ex_2015[i], net_2015[i], 2015])
data_2015 = pd.DataFrame(data_2015, columns=['Country',  'exports','net_loans', 'year'])

data_2016 = []

ex_2016 = exports['2016'].tolist()
net_2016 = net_loans['2016'].tolist()
for i in range(len(country)):
    data_2016.append([country[i], ex_2016[i], net_2016[i], 2016])
data_2016 = pd.DataFrame(data_2016, columns=['Country', 'exports','net_loans', 'year'])

data_2017 = []

ex_2017 = exports['2017'].tolist()
net_2017 = net_loans['2017'].tolist()
for i in range(len(country)):
    data_2017.append([country[i],ex_2017[i], net_2017[i], 2017])
data_2017 = pd.DataFrame(data_2017, columns=['Country', 'exports','net_loans', 'year'])

data_2018 = []

ex_2018 = exports['2018'].tolist()
net_2018 = net_loans['2018'].tolist()
for i in range(len(country)):
    data_2018.append([country[i],ex_2018[i], net_2018[i], 2018])
data_2018 = pd.DataFrame(data_2018, columns=['Country', 'exports', 'net_loans', 'year'])

data_2019 = []
ex_2019 = exports['2019'].tolist()
net_2019 = net_loans['2019'].tolist()
for i in range(len(country)):
    data_2019.append([country[i], ex_2019[i], net_2019[i], 2019])
data_2019 = pd.DataFrame(data_2019, columns=['Country', 'exports', 'net_loans', 'year'])

data_2020 = []

ex_2020 = exports['2020'].tolist()
net_2020 = net_loans['2020'].tolist()
for i in range(len(country)):
    data_2020.append([country[i], ex_2020[i],net_2020[i], 2020])
data_2020 = pd.DataFrame(data_2020, columns=['Country', 'exports', 'net_loans', 'year'])

dataset_clean = pd.concat([data_2012, data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019, data_2020], ignore_index=True)
# dataset_clean = dataset_clean.groupby('Country')
# dataset_clean = dataset_clean.filter(lambda full_data_clean: full_data_clean.notnull().values.all())
# dataset_clean = dataset_clean.groupby('Country').filter(lambda x: x['year'].count() == 9)
dataset_clean
```

```python
# read CPI dataset
CPI = pd.read_csv("final dataset/CPI2021_GlobalResults_Trends.csv", header = 0)
# restracture CPI dataset
CPI_2012 = CPI[['Country','CPI Score 2012']]
CPI_2012.rename(columns = {'CPI Score 2012': 'CPI'}, inplace = True)
CPI_2012['year'] = 2012

CPI_2013 = CPI[['Country','CPI Score 2013']]
CPI_2013.rename(columns = {'CPI Score 2013': 'CPI'}, inplace = True)
CPI_2013['year'] = 2013

CPI_2014 = CPI[['Country','CPI score 2014']]
CPI_2014.rename(columns = {'CPI score 2014': 'CPI'}, inplace = True)
CPI_2014['year'] = 2014

CPI_2015 = CPI[['Country','CPI score 2015']]
CPI_2015.rename(columns = {'CPI score 2015': 'CPI'}, inplace = True)
CPI_2015['year'] = 2015

CPI_2016 = CPI[['Country','CPI score 2016']]
CPI_2016.rename(columns = {'CPI score 2016': 'CPI'}, inplace = True)
CPI_2016['year'] = 2016

CPI_2017 = CPI[['Country','CPI score 2017']]
CPI_2017.rename(columns = {'CPI score 2017': 'CPI'}, inplace = True)
CPI_2017['year'] = 2017

CPI_2018 = CPI[['Country','CPI score 2018']]
CPI_2018.rename(columns = {'CPI score 2018': 'CPI'}, inplace = True)
CPI_2018['year'] = 2018

CPI_2019 = CPI[['Country','CPI score 2019']]
CPI_2019.rename(columns = {'CPI score 2019': 'CPI'}, inplace = True)
CPI_2019['year'] = 2019

CPI_2020 = CPI[['Country','CPI score 2020']]
CPI_2020.rename(columns = {'CPI score 2020': 'CPI'}, inplace = True)
CPI_2020['year'] = 2020

CPI_clean = pd.concat([CPI_2012,CPI_2013,CPI_2014,CPI_2015,CPI_2016,CPI_2017,CPI_2018,CPI_2019,CPI_2020], ignore_index=True)
# CPI_clean = CPI_clean.groupby('Country')
# CPI_clean = CPI_clean.filter(lambda full_data_clean: full_data_clean.notnull().values.all())
# CPI_clean = CPI_clean.groupby('Country').filter(lambda x: x['year'].count() == 9)
```

```python
# read economics freedom data
eco_free = pd.read_csv("final dataset/efotw-2022-master-index-data-for-researchers-iso.csv", header = 0)
eco_free_clean = eco_free[['Country', 'year', 'Size of Government', 'Legal System & Property Rights', 'Sound Money', 'Freedom to trade internationally', 'Regulation']]
# read unemployment rate data
unemployment_rate = pd.read_csv("final dataset/unemployment-rate(1).csv", header = 0)
# inflation = pd.read_csv("final dataset/annual-inflation-of-consumer-prices(1).csv", header = 0)
# part_merge = pd.merge(unemployment_rate,inflation,how='inner',on=['Country','year'])
#merge datasets
merge = pd.merge(CPI_clean,unemployment_rate,how='inner',on=['Country','year'])
first_merge = pd.merge(merge,eco_free_clean,how='inner',on=['Country','year'])
second_data = pd.merge(first_merge,dataset_clean,how='inner',on=['Country','year'])
full_data = pd.merge(second_data,dt_all,how='inner',on=['Country','year'])
```
